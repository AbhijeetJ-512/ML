### Importing Librarires


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import random
```

Reading the data from a .csv file


```python
path='data/student_marks.csv'
data= pd.read_csv(path)
data
```

### Dropping the student ID because of no requiremnet for analysis

And adding a mean column to the data set


```python
marks = data.drop(['Student_ID'],axis=1)
marks['mean_score']=(marks.Test_1+marks.Test_2+marks.Test_3+marks.Test_4+marks.Test_5+marks.Test_6+marks.Test_7+marks.Test_8+marks.Test_9+marks.Test_10+marks.Test_11+marks.Test_12)/12
marks
```

Plotting the graph for the average score and finding its mean median mode in graph


```python
marks["mean_score"].plot.density(label='Mean_score of all')
plt.axvline(x=marks['mean_score'].mean(),color='r',label='mean',linestyle='--')
plt.axvline(x=marks['mean_score'].median(),color='b',label='median',linestyle='--')
plt.axvline(x=marks['mean_score'].mode()[0],color='g',label='mode',linestyle='--')
plt.legend()
plt.show()
```

The ROCK PAPER SCISCOR GAME


```python
n=int(input("enter no of turns"))
count=np.zeros((3,))
comp_i=[]
user_i=[]
result=[]
gamelist=['rock','paper','sisccor']
for x in range(n):
    comp=random.choice(gamelist)
    comp_i.append(comp)
    user=input("Rock paper or sisccor?")
    user_i.append(user)
    if(comp=='rock' and user=='paper') or (comp=='paper' and user=='sisccor') or (comp=='sisccor' and user=='rock'):
        print("YOU WIN")
        count[0]+=1
        result.append('win')
    elif(comp==user):
        print("TIE")
        count[1]+=1
        result.append('tie')
    else:
        print("Computer WINS!")
        count[2]+=1
        result.append('loss')
print(f"computer won : {count[2]} time(s)\nYou won : {count[0]} time(s)\nTie : {count[1]} time(s)")
```

Coverting the data into a Dataframe and saving it as a .csv file


```python
data={'computer':comp_i,'user':user_i,'result':result}
df=pd.DataFrame(data)
df.to_csv('data/game.csv')
```

PLottting the data in pie chart and giving the percentage of win rate


```python
plt.pie(count,labels=['User','Tie','Computer'],autopct='%1.1f%%')
plt.title('Pie plot')
plt.legend(loc='upper left')
plt.show()
```
