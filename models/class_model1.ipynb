```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path ="data/Uni_linear.txt"
data = pd.read_csv(path,header=None)
data
```

```python
plt.scatter(data[0],data[1],marker='x',label='given data')
plt.title("Data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x_data = data[0].to_numpy().reshape(-1,1)
y_data = data[1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_squared_error
model = LinearRegression()
model.fit(x_data,y_data)
```

```python
predict = model.predict(x_data)
error = mean_squared_error(predict,y_data)
print(error)
model.score(x_data,y_data)
```

```python
plt.scatter(data[0],data[1],marker='x',label='given data')
plt.plot(data[0],predict,label='predicited',color='r')
plt.title("Data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

MODel polynomial


```python
from sklearn.preprocessing import PolynomialFeatures
degree=5
poly = PolynomialFeatures(degree=degree)
x_data = poly.fit_transform(x_data)
model1=LinearRegression()
model1.fit(x_data,y_data)
model1.score(x_data,y_data)
```
