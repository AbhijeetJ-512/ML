```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  log_loss
```

```python
path='data/tumor.csv'
data = pd.read_csv(path)
data.columns
data = data.sort_values(by='size')
```

```python
tumor_data=data[data.tumor==1]
no_tumor_data=data[data.tumor==0]
plt.scatter(tumor_data['size'], tumor_data['tumor'], marker='x', color='red')
plt.scatter(no_tumor_data['size'], no_tumor_data['tumor'], marker='o', color='blue')
plt.show()
```

```python
x_data = np.array(data['size']).reshape(-1,1)
y_data = np.array(data['tumor'])
```

```python
model = LogisticRegression()
model.fit(x_data,y_data)
```

```python
prediction = model.predict(x_data)
tumor_data=data[data.tumor==1]
no_tumor_data=data[data.tumor==0]
plt.scatter(tumor_data['size'], tumor_data['tumor'], marker='x', color='red')
plt.scatter(no_tumor_data['size'], no_tumor_data['tumor'], marker='o', color='blue')
plt.plot(x_data,prediction,color='g')
plt.show()
```

```python
error =log_loss(y_data,prediction)
print(error)
value=model.predict([[0.8]])
print(value)
```
