```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
path = 'data/x_y_excel.csv'
data=pd.read_csv(path)
print(data)
x_data=np.array([data.x_data]).reshape(-1,1)
y_data=np.array(data.y_data)
print(x_data)
```

```python
model = LinearRegression()
model.fit(x_data,y_data)
plt.scatter(x_data,y_data,marker='x')
plt.xlabel('x axis')
plt.ylabel("y axis")
plt.show()
```

```python
prediction = model.predict(x_data)
plt.scatter(x_data,y_data,marker='x',color='violet',label='real')
plt.plot(x_data,prediction,linestyle=':',color='green',label='prediction')
plt.xlabel('x axis')
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(prediction,y_data)
print(f"error {mse}")
print(f"intercept {model.intercept_} and bias {model.coef_}")
```
